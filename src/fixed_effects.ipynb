```r
fe_twoway_full <- plm(log_death_rate ~ fent_gms + oxy_gms + hydro_gms + 
                      unempl_rate + poverty_rate + snap_rate + 
                      medicaid_rate + gsp_per_cap + min_wage, 
                      data = panel_df, 
                      model = "within", 
                      effect = "twoways")

summary(fe_twoway_full)
```

```r
# We use lag() on the drugs and the economic drivers
fe_lag_full <- plm(log_death_rate ~ lag(fent_gms, 1) + lag(oxy_gms, 1) + lag(hydro_gms, 1) + 
                   lag(unempl_rate, 1) + lag(poverty_rate, 1) + lag(snap_rate, 1) + 
                   medicaid_rate + gsp_per_cap + min_wage, 
                   data = panel_df, 
                   model = "within", 
                   effect = "twoways")

summary(fe_lag_full)
```
